```python
import numpy as np
import matplotlib.pyplot as plt

Radius = 4.0
radius = 1.0

wplanet = 2.0
wmoon = 14.0

time = np.arange(0.0, 3.2, 0.02)

xo = Radius * np.cos(wplanet*time)
yo = Radius * np.sin(wplanet*time)

x = Radius *np.cos(wplanet*time) +  radius *np.cos(wmoon*time)

y = Radius *np.sin(wplanet*time) +  radius *np.sin(wmoon*time)

plt.plot(x,y, label = 'Moon')
plt.plot(xo,yo, label = 'Planet')

plt.title('Motion of Earth and Moon')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np


t = np.linspace(0, 2*np.pi, 1000)


x_earth = np.cos(t)
y_earth = np.sin(t)


x_moon = x_earth + 0.1*np.cos(13*t)
y_moon = y_earth + 0.1*np.sin(13*t)

plt.figure(figsize=(6,6))


plt.plot(x_earth, y_earth, label='Earth\'s orbit around the Sun')


plt.plot(x_moon, y_moon, label='Moon\'s orbit around the Earth')

plt.legend()
plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


t = np.linspace(0, 2*np.pi, 1000)


x_earth = np.cos(t)
y_earth = np.sin(t)


x_moon = x_earth + 0.1*np.cos(13*t)
y_moon = y_earth + 0.1*np.sin(13*t)

x_c = x_moon + 0.04*np.cos(200*t)
y_c = y_moon + 0.04*np.sin(200*t)

plt.figure(figsize=(6,6))


plt.plot(x_earth, y_earth, label='Earth\'s orbit around the Sun')


plt.plot(x_moon, y_moon, label='Moon\'s orbit around the Earth')

plt.plot(x_c, y_c, label='Chandrayaan\'s orbit around the Moon')

plt.legend()
plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

t = np.linspace(0, 2*np.pi, 1000)

x_earth = t
y_earth = np.zeros_like(t)

x_moon = x_earth
y_moon = 0.1 * np.sin(13*t)

plt.figure(figsize=(6,6))

plt.plot(x_earth, y_earth, label='Earth\'s path without the Sun')

plt.plot(x_moon, y_moon, label='Moon\'s path around the Earth')

plt.xlim(-2,7)
plt.ylim(-2,3)

plt.legend()
plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
