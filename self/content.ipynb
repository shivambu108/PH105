```python
import pandas as pd

# Define the path to your CSV file
csv_file_path = r"C:\Users\shiva\Downloads\netflix-report\CONTENT_INTERACTION\ViewingActivity.csv"

# Define the path for the output file
output_file_path = r"C:\Users\shiva\Downloads\netflix-report\CONTENT_INTERACTION\S_Watch_History.csv"

# Define the user to filter by
target_user = "Shivambu"

# Read the CSV file into a DataFrame
df = pd.read_csv(csv_file_path)

# Filter the DataFrame by the target user
user_data = df[df['Profile Name'] == target_user]

# Select the 'Title' column
watch_history = user_data[['Title']]

# Save the extracted titles to a new CSV file
watch_history.to_csv(output_file_path, index=False)

print(f"Watch history for {target_user} has been saved to {output_file_path}")
```
