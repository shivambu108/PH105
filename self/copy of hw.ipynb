### Simple Harmonic Oscillator (SHO):

1. **Definition**:
   - A simple harmonic oscillator is a system that exhibits harmonic motion around an equilibrium position under the influence of a restoring force proportional to its displacement from the equilibrium.
   - It is a fundamental model in physics and describes various real-world systems, such as mass-spring systems, pendulums, and vibrating strings.

2. **Equation of Motion**:
   - The motion of a simple harmonic oscillator is governed by a second-order linear ordinary differential equation (ODE).
   - This equation, derived from Newton's Second Law and Hooke's Law, relates the acceleration of the object to its displacement:
     $$ m \frac{d^2x}{dt^2} + kx = 0 $$


   - Here, \( m \) is the mass of the object, \( k \) is the spring constant, \( x \) is the displacement from equilibrium, and \( t \) is time.

3. **Restoring Force**:
   - The restoring force acting on the oscillator is given by Hooke's Law:
     $$ F_{\text{restoring}} = -kx $$
   - This force is proportional to the displacement from the equilibrium position \( x \) and acts in the opposite direction to the displacement.
   - It is responsible for restoring the oscillator back towards equilibrium when displaced.

4. **Periodic Motion**:
   - Due to the restoring force, the oscillator undergoes periodic motion characterized by a sinusoidal pattern.
   - The motion repeats at regular intervals, and the displacement oscillates back and forth around the equilibrium position.

5. **Equilibrium Position**:
   - The equilibrium position is the position where the net force acting on the object is zero, and the object is at rest.
   - The equilibrium position serves as the reference point for the oscillator's motion. Any displacement from this position results in a restoring force that drives the oscillator back towards equilibrium. Understanding the behavior of the oscillator around the equilibrium position is crucial for analyzing its stability and response to external perturbations.

6. **Frequency and Period**:
   - The frequency (f) and period (T) of the oscillator represent how quickly it oscillates back and forth around the equilibrium position.
   - The frequency and period are inversely related and depend on the mass of the object and the stiffness of the spring. They provide important information about the time it takes for one complete oscillation and the number of oscillations per unit time. These characteristics are fundamental for describing the dynamic behavior of the oscillator and are often used to characterize its performance in various applications.

7. **Energy Conservation**:
   - In an ideal SHO without external damping forces, mechanical energy (kinetic energy + potential energy) is conserved.
   - This conservation of energy ensures that the oscillator's motion remains stable over time. The interchange between kinetic and potential energy as the oscillator oscillates ensures that its total energy remains constant. Understanding energy conservation allows for the prediction of the oscillator's behavior and the analysis of its long-term dynamics.

8. **Phase and Phase Constant**:
   - The phase of the oscillator represents its position within one complete cycle of oscillation, measured in radians or degrees.
   - The phase provides information about the timing and relationship between different oscillatory systems. The phase constant (\phi) in the solution of the SHO equation represents the initial phase of the oscillator at t = 0. It determines the starting point of the oscillator's motion within its oscillatory cycle. Understanding phase relationships is crucial for analyzing the synchronization and interference of oscillatory systems in various applications.



### Solution to the Differential Equation:
The solution to this differential equation yields the equation of motion of the SHO, which is:

$$ x(t) = A \sin(\omega t + \phi) $$

Where:
- \( A \) is the amplitude of the oscillation.
- \( omega \) is the angular frequency of the oscillation: $$ \omega = \sqrt{\frac{k}{m}} $$
- \( phi \) is the phase constant, representing the initial phase of the oscillation at \( t = 0 \).

### Interpretation:
- It Describes the position (\( x \)) of the object as a function of time (\( t \)).
- Represents harmonic motion, where the object oscillates sinusoidally about the equilibrium position.
- Amplitude (\( A \)), angular frequency (\( \omega \)), and phase constant (\( \phi \)) determine the characteristics of the oscillatory motion.

Understanding the differential equation of the SHO is crucial for analyzing its behavior, predicting its motion, and designing systems that utilize simple harmonic motion.



### Numerical Simulation:

### Euler's Method
Euler's method is a basic numerical technique used to approximate solutions to ordinary differential equations (ODEs). Here's an organized overview of the method:

1. **Discretization of Time**:
   - Time domain is divided into small time steps (\( \Delta t \)).

2. **Initial Conditions**:
   - Initial value of the function (\( x(0) \)) and its derivative (\( \frac{dx}{dt}(0) \)) are provided.

3. **Iterative Approximation**:
   - Starting from the initial conditions, the value of the function at each time step is approximated using the derivative at the previous time step.

4. **Update Rule**:
   - The update rule is:
     $$ x_{i+1} = x_i + \frac{dx}{dt}_i \cdot \Delta t $$

5. **Repeat**:
   - Steps 3 and 4 are repeated until the desired time is reached.

### Euler-Cromer's Method
Euler-Cromer's method is a variation of Euler's method that improves the accuracy of the approximation by updating the velocity before updating the position. Here's how it differs:

1. **Initial Conditions**:
   - Same as Euler's method.

2. **Update Rule**:
   - The update rules are:
     $$ v_{i+1} = v_i + \frac{dv}{dt}_i \cdot \Delta t $$
     $$ x_{i+1} = x_i + v_{i+1} \cdot \Delta t $$

3. **Repeat**:
   - Same as Euler's method.

### Implementation
In the provided code:
- Both Euler's method and Euler-Cromer's method are implemented as functions `euler_method` and `euler_cromer_method`, respectively.
- These functions take initial conditions (amplitude, angular frequency, time step, number of steps) as input parameters and return arrays of time values, positions, and velocities.
- Both methods are used to simulate the motion of a simple harmonic oscillator and compare their results with the exact solution obtained from the differential equation of motion.



```python
import numpy as np
import matplotlib.pyplot as plt

# Define the spring constant and mass
k = 1  # N/m
m = 1  # kg

# Define the equations of motion
def displacement(t, A, omega):
    return A * np.sin(omega * t)

def velocity(t, A, omega):
    return A * omega * np.cos(omega * t)

def acceleration(x, k, m):
    return -k/m * x

# Define Euler's method for numerical integration
def euler_method(A, omega, dt, num_steps):
    t_values = np.linspace(0, dt * (num_steps - 1), num_steps)
    x_values = np.zeros(num_steps)
    v_values = np.zeros(num_steps)
    
    x_values[0] = displacement(0, A, omega)
    v_values[0] = velocity(0, A, omega)
    
    for i in range(1, num_steps):
        v_values[i] = v_values[i-1] + acceleration(x_values[i-1], k, m) * dt
        x_values[i] = x_values[i-1] + v_values[i-1] * dt
    
    return t_values, x_values, v_values

# Define Euler-Cromer's method for numerical integration
def euler_cromer_method(A, omega, dt, num_steps):
    t_values = np.linspace(0, dt * (num_steps - 1), num_steps)
    x_values = np.zeros(num_steps)
    v_values = np.zeros(num_steps)
    
    x_values[0] = displacement(0, A, omega)
    v_values[0] = velocity(0, A, omega)
    
    for i in range(1, num_steps):
        v_values[i] = v_values[i-1] + acceleration(x_values[i-1], k, m) * dt
        x_values[i] = x_values[i-1] + v_values[i] * dt  # Use updated velocity here
    
    return t_values, x_values, v_values



# Initial conditions
A = 1  # Amplitude in meters
omega = 1  # Angular frequency in rad/s
dt = 0.1  # Time step size
num_steps = 1000  # Number of time steps

# Exact solution
t_exact = np.linspace(0, dt * (num_steps - 1), num_steps)
x_exact = displacement(t_exact, A, omega)
v_exact = velocity(t_exact, A, omega)

# Using Euler's method
t_euler, x_euler, v_euler = euler_method(A, omega, dt, num_steps)

# Using Euler-Cromer's method
t_ec, x_ec, v_ec = euler_cromer_method(A, omega, dt, num_steps)

# Calculate errors
errors_euler = np.abs(x_exact - x_euler)
errors_ec = np.abs(x_exact - x_ec)

# Plotting
fig, (ax1, ax2,ax3,ax4) = plt.subplots(4, 1, figsize=(10, 20))

# Plotting Euler's method with larger, red circle markers
ax1.plot(t_euler, x_euler, label="Euler's Method", linestyle='dashed', marker='^', markersize=5, color='red', alpha=0.7)

# Plotting Euler-Cromer's method with larger, green square markers
ax1.plot(t_ec, x_ec, label="Euler-Cromer's Method", linestyle='dashed', marker='^', markersize=3, color='green')

# Plotting original growth curve with a solid black line
ax1.plot(t_exact, x_exact, label='Exact', color='black', linewidth=2)

ax1.set_xlabel('Time (s)')
ax1.set_ylabel('Displacement (m)')
ax1.set_title('Displacement vs Time for Simple Harmonic Oscillator')
ax1.legend()
ax1.grid(True)

# Zoomed plot 1
zoom_region = (45, 50)  # Define the region you want to zoom in on
ax2.plot(t_euler, x_euler, label="Euler's Method", linestyle='dashed', marker='^', markersize=5, color='red', alpha=0.7)
ax2.plot(t_ec, x_ec, label="Euler-Cromer's Method", linestyle='dashed', marker='^', markersize=3, color='green')
ax2.plot(t_exact, x_exact, label='Exact', color='black', linewidth=2)
ax2.set_xlim(*zoom_region)
ax2.set_ylim(-20,20)
ax2.set_xlabel('Time (s)')
ax2.set_ylabel('Displacement (m)')
ax2.set_title('Zoomed In (1) Displacement vs Time')
ax2.legend()
ax2.grid(True)


# Zoomed plot 2
zoom_region = (95, 100)  # Define the region you want to zoom in on
ax3.plot(t_euler, x_euler, label="Euler's Method", linestyle='dashed', marker='^', markersize=5, color='red', alpha=0.7)
ax3.plot(t_ec, x_ec, label="Euler-Cromer's Method", linestyle='dashed', marker='^', markersize=3, color='green')
ax3.plot(t_exact, x_exact, label='Exact', color='black', linewidth=2)
ax3.set_xlim(*zoom_region)
ax3.set_ylim(-150,150)
ax3.set_xlabel('Time (s)')
ax3.set_ylabel('Displacement (m)')
ax3.set_title('Zoomed In (2) Displacement vs Time')
ax3.legend()
ax3.grid(True)



# Plotting errors

ax4.plot(t_euler, errors_euler, label="Euler's Method Error", linestyle='dashed', marker='^', markersize=5, color='red', alpha=0.7)
ax4.plot(t_ec, errors_ec, label="Euler-Cromer's Method Error",linestyle='-.', marker='^', markersize=2, color='green')
ax4.set_xlabel('Time (s)')
ax4.set_ylabel('Error (m)')
ax4.set_title('Error in Displacement Approximation')
ax4.legend()
ax4.grid(True)

plt.tight_layout()
plt.show()

# Display table
print("Time  | Exact Displacement | Euler's Method  | Euler-Cromer's Method  | Euler's Method Error | Euler-Cromer's Method Error")
print("-" * 160)
for i in range(len(t_euler)):
    print("{:.2f}  |  {:.4f}  |  {:.4f}  |  {:.4f}  |  {:.4f}  |  {:.4f}".format(t_euler[i], x_exact[i], x_euler[i], x_ec[i], errors_euler[i], errors_ec[i]))

```
