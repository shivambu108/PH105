## Triple Pendulum Equations of Motion

### Kinetic Energy \( T \)

\[
\begin{aligned}
T &= \frac{1}{2} m_1 l_1^2 \dot{\theta}_1^2 \\
  &+ \frac{1}{2} m_2 \left( l_1^2 \dot{\theta}_1^2 + l_2^2 \dot{\theta}_2^2 + 2 l_1 l_2 \cos(\theta_1 - \theta_2) \dot{\theta}_1 \dot{\theta}_2 \right) \\
  &+ \frac{1}{2} m_3 \left( l_1^2 \dot{\theta}_1^2 + l_2^2 \dot{\theta}_2^2 + l_3^2 \dot{\theta}_3^2 + 2 l_1 l_2 \cos(\theta_1 - \theta_2) \dot{\theta}_1 \dot{\theta}_2 + 2 l_1 l_3 \cos(\theta_1 - \theta_3) \dot{\theta}_1 \dot{\theta}_3 + 2 l_2 l_3 \cos(\theta_2 - \theta_3) \dot{\theta}_2 \dot{\theta}_3 \right)
\end{aligned}
\]

### Potential Energy \( V \)

\[
V = -g \left( m_1 l_1 \cos\theta_1 + m_2 (l_1 \cos\theta_1 + l_2 \cos\theta_2) + m_3 (l_1 \cos\theta_1 + l_2 \cos\theta_2 + l_3 \cos\theta_3) \right)
\]

### Lagrangian \( L \)

\[
L = T - V
\]

### Euler-Lagrange Equations

#### For \( \theta_1 \)

\[
\begin{aligned}
\frac{\partial L}{\partial \dot{\theta}_1} &= m_1 l_1^2 \dot{\theta}_1 + m_2 \left( l_1^2 \dot{\theta}_1 + l_1 l_2 \cos(\theta_1 - \theta_2) \dot{\theta}_2 \right) + m_3 \left( l_1^2 \dot{\theta}_1 + l_1 l_2 \cos(\theta_1 - \theta_2) \dot{\theta}_2 + l_1 l_3 \cos(\theta_1 - \theta_3) \dot{\theta}_3 \right) \\
\frac{d}{dt} \left( \frac{\partial L}{\partial \dot{\theta}_1} \right) &= m_1 l_1^2 \ddot{\theta}_1 + m_2 \left( l_1^2 \ddot{\theta}_1 + l_1 l_2 \left( \cos(\theta_1 - \theta_2) \ddot{\theta}_2 - \sin(\theta_1 - \theta_2) (\dot{\theta}_1 - \dot{\theta}_2) \dot{\theta}_2 \right) \right) \\
&\quad + m_3 \left( l_1^2 \ddot{\theta}_1 + l_1 l_2 \left( \cos(\theta_1 - \theta_2) \ddot{\theta}_2 - \sin(\theta_1 - \theta_2) (\dot{\theta}_1 - \dot{\theta}_2) \dot{\theta}_2 \right) + l_1 l_3 \left( \cos(\theta_1 - \theta_3) \ddot{\theta}_3 - \sin(\theta_1 - \theta_3) (\dot{\theta}_1 - \dot{\theta}_3) \dot{\theta}_3 \right) \right) \\
\frac{\partial L}{\partial \theta_1} &= g \left( m_1 l_1 \sin\theta_1 + m_2 (l_1 \sin\theta_1) + m_3 (l_1 \sin\theta_1) \right)
\end{aligned}
\]

\[
\begin{aligned}
& m_1 l_1^2 \ddot{\theta}_1 + m_2 \left( l_1^2 \ddot{\theta}_1 + l_1 l_2 \left( \cos(\theta_1 - \theta_2) \ddot{\theta}_2 - \sin(\theta_1 - \theta_2) (\dot{\theta}_1 - \dot{\theta}_2) \dot{\theta}_2 \right) \right) \\
& + m_3 \left( l_1^2 \ddot{\theta}_1 + l_1 l_2 \left( \cos(\theta_1 - \theta_2) \ddot{\theta}_2 - \sin(\theta_1 - \theta_2) (\dot{\theta}_1 - \dot{\theta}_2) \dot{\theta}_2 \right) + l_1 l_3 \left( \cos(\theta_1 - \theta_3) \ddot{\theta}_3 - \sin(\theta_1 - \theta_3) (\dot{\theta}_1 - \dot{\theta}_3) \dot{\theta}_3 \right) \right) \\
& - g \left( m_1 l_1 \sin\theta_1 + m_2 (l_1 \sin\theta_1) + m_3 (l_1 \sin\theta_1) \right) = 0
\end{aligned}
\]

#### For \( \theta_2 \)

\[
\begin{aligned}
\frac{\partial L}{\partial \dot{\theta}_2} &= m_2 \left( l_2^2 \dot{\theta}_2 + l_1 l_2 \cos(\theta_1 - \theta_2) \dot{\theta}_1 \right) + m_3 \left( l_2^2 \dot{\theta}_2 + l_2 l_3 \cos(\theta_2 - \theta_3) \dot{\theta}_3 + l_1 l_2 \cos(\theta_1 - \theta_2) \dot{\theta}_1 \right) \\
\frac{d}{dt} \left( \frac{\partial L}{\partial \dot{\theta}_2} \right) &= m_2 \left( l_2^2 \ddot{\theta}_2 + l_1 l_2 \left( \cos(\theta_1 - \theta_2) \ddot{\theta}_1 - \sin(\theta_1 - \theta_2) (\dot{\theta}_1 - \dot{\theta}_2) \dot{\theta}_1 \right) \right) \\
&\quad + m_3 \left( l_2^2 \ddot{\theta}_2 + l_2 l_3 \left( \cos(\theta_2 - \theta_3) \ddot{\theta}_3 - \sin(\theta_2 - \theta_3) (\dot{\theta}_2 - \dot{\theta}_3) \dot{\theta}_3 \right) + l_1 l_2 \left( \cos(\theta_1 - \theta_2) \ddot{\theta}_1 - \sin(\theta_1 - \theta_2) (\dot{\theta}_1 - \dot{\theta}_2) \dot{\theta}_1 \right) \right) \\
\frac{\partial L}{\partial \theta_2} &= g \left( m_2 l_2 \sin\theta_2 + m_3 (l_2 \sin\theta_2) \right)
\end{aligned}
\]

\[
\begin{aligned}
& m_2 \left( l_2^2 \ddot{\theta}_2 + l_1 l_2 \left( \cos(\theta_1 - \theta_2) \ddot{\theta}_1 - \sin(\theta_1 - \theta_2) (\dot{\theta}_1 - \dot{\theta}_2) \dot{\theta}_1 \right) \right) \\
& + m_3 \left( l_2^2 \ddot{\theta}_2 + l_2 l_3 \left( \cos(\theta_2 - \theta_3) \ddot{\theta}_3 - \sin(\theta_2 - \theta_3) (\dot{\theta}_2 - \dot{\theta}_3) \dot{\theta}_3 \right) + l_1 l_2 \left( \cos(\theta_1 - \theta_2) \ddot{\theta}_1 - \sin(\theta_1 - \theta_2) (\dot{\theta}_1 - \dot{\theta}_2) \dot{\theta}_1 \right) \right) \\
& - g \left( m_2 l_2 \sin\theta_2 + m_3 (l_2 \sin\theta_2) \right) = 0
\end{aligned}
\]

#### For \( \theta_3 \)

\[
\begin{aligned}
\frac{\partial L}{\partial \dot{\theta}_3} &= m_3 \left( l_3^2 \dot{\theta}_3 + l_2 l_3 \cos(\theta_2 - \theta_3) \dot{\theta}_2 + l_1 l_3 \cos(\theta_1 - \theta_3) \dot{\theta}_1 \right) \\
\frac{d}{dt} \left( \frac{\partial L}{\partial \dot{\theta}_3} \right) &= m_3 \left( l_3^2 \ddot{\theta}_3 + l_2 l_3 \left( \cos(\theta_2 - \theta_3) \ddot{\theta}_2 - \sin(\theta_2 - \theta_3) (\dot{\theta}_2 - \dot{\theta}_3) \dot{\theta}_2 \right) + l_1 l_3 \left( \cos(\theta_1 - \theta_3) \ddot{\theta}_1 - \sin(\theta_1 - \theta_3) (\dot{\theta}_1 - \dot{\theta}_3) \dot{\theta}_1 \right) \right) \\
\frac{\partial L}{\partial \theta_3} &= g m_3 l_3 \sin\theta_3
\end{aligned}
\]

\[
\begin{aligned}
& m_3 \left( l_3^2 \ddot{\theta}_3 + l_2 l_3 \left( \cos(\theta_2 - \theta_3) \ddot{\theta}_2 - \sin(\theta_2 - \theta_3) (\dot{\theta}_2 - \dot{\theta}_3) \dot{\theta}_2 \right) + l_1 l_3 \left( \cos(\theta_1 - \theta_3) \ddot{\theta}_1 - \sin(\theta_1 - \theta_3) (\dot{\theta}_1 - \dot{\theta}_3) \dot{\theta}_1 \right) \right) \\
& - g m_3 l_3 \sin\theta_3 = 0
\end{aligned}
\]
